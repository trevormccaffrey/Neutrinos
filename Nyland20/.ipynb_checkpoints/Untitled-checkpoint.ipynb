```python
#import necessary packages
from astropy.table import Table
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from astropy.stats import sigma_clip
#Default MPL colorcycle
mplcol = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

Load in All-sky Point-source catalog from IceCube.  Most of these "events" will just be from the Earth's atmosphere, but we may be able to separate astrophysical neutrinos and associate them with individual events (such as jets launching between FIRST and VLASS timescales, Nyland+2020).

```python
tbl1 = Table.read("../ICall_2008-2018_exp.fits")
ICall = tbl1.to_pandas()
ICall.head()
```

Get a general feel for the IceCube dataset.

```python
ICall.iloc[:,0]
```

```python
for i in range(ICall.shape[1]):
    fig = plt.figure(figsize=(6,6))
    plt.hist(ICall.iloc[:,i], bins="scott")
    plt.xlabel(ICall.columns[i])
    plt.show()
```

* Increases in total events up to given MJDs because the number of vertical strings (detectors) increases with epoch
* Energies of events is clearly bimodal -- probably showing the separation of atmospheric/astrophysical neutrinos
* Positions of events have angular errors ≥0.2deg, with a heavy tail out to $\sim$14deg
* Events at zenith an near the equator are favored since we're at the south pole

Now load in a match of this file to the Nyland+2020 sources detected in SDSS-DR16

```python
tbl2 = Table.read("ICmatches/Nyland20_IceCube_0.5deg.fits")
IC_match = tbl2.to_pandas()
print(IC_match.shape)
IC_match.head()
```

```python
for i in range(7, IC_match.shape[1]):
    fig = plt.figure(figsize=(6,6))
    plt.hist(IC_match.iloc[:,i], bins="scott")
    plt.xlabel(IC_match.columns[i])
    plt.show()
```

```python
names = IC_match["SDSS_NAME"].values
ra = IC_match["RA"].values
dec = IC_match["DEC"].values
E = IC_match["log10(E/GeV)"].values
```

```python
np.unique(names)
```

```python
def print_events(name):
    print("Number of events for %s: %d" % (str(name)[2:-1], (names==name).sum()))
for i in np.unique(names):
    print_events(i)
```

```python
df_0742 = IC_match[names==b'074248.40+270412.2']
df_0807 = IC_match[names==b'080704.32+210208.6']
df_0832 = IC_match[names==b'083247.64+230234.1']
df_1157 = IC_match[names==b'115733.23+314227.0']
df_1447 = IC_match[names==b'144716.22+051226.5']
df_1603 = IC_match[names==b'160318.74+180904.9']
```

```python
def plot_light_curve(df):
    fig = plt.figure(figsize=(8,4))
    #plt.scatter(df["MJD[days]"].values, 10.**(df["log10(E/GeV)"].values)/1000)
    plt.scatter(df["MJD[days]"].values, df["log10(E/GeV)"].values)
    xlims = plt.gca().get_xlim()
    plt.plot([50000,60000],[np.log10(1000*23.3),np.log10(1000*23.3)],"--g",label="TXS 0506+056")
    #plt.plot([50000,60000],[23.3,23.3],"--g",label="TXS 0506+056")
    plt.xlim(xlims)
    plt.title(str(df["SDSS_NAME"].values[0])[2:-1])
    plt.ylabel("log10(E/GeV)")
    #plt.ylabel("TeV")
    plt.xlabel("MJD [days]")
    plt.legend(loc="upper right")
    plt.show()
```

```python
plot_light_curve(df_0742)
```

```python
plot_light_curve(df_0807)
```
